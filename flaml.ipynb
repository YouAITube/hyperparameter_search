```python
import numpy as np
from flaml import AutoML
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python

df = load_wine(as_frame=True)
df["data"].head(2)
```

```python
df['DESCR']
```

```python
X, y = final_data=df['data'], df.target
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
automl = AutoML()

parametrs = {
     "time_budget": 120,
     "metric": "accuracy",
     "task": "classification"
}
```

```python
automl.fit(X_train=X_train, y_train=y_train, **parametrs)
```

```python
pred_prob = automl.predict_proba(X_test)
pred_prob_shape = automl.predict_proba(X_test).shape
best_model = automl.model
print(pred_prob_shape)
print(best_model)
```

```python
from sklearn.metrics import accuracy_score as acc
acc(y_test,np.argmax(pred_prob,axis=1) )
```

```python
import pandas as pd

result = pd.DataFrame({
    'target': y_test,
    'prediction': np.argmax(pred_prob,axis=1),
})

result
```

```python
import pickle

with open("automl.pkl", "wb") as f:
    pickle.dump(automl, f, pickle.HIGHEST_PROTOCOL)

# At prediction time
with open("automl.pkl", "rb") as f:
    automl = pickle.load(f)
pred = automl.predict(X_test)
```
